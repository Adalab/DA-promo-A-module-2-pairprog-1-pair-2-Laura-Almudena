```python
# Vamos con la columna fatal_(y/n). Lo primero que tenéis que hacer es evaluar los valores únicos que hay. 
# Esperaríamos tener solo dos valores: y y n, pero tristemente no es así. 
# Cread una columna nueva y una función que devuelva unicamente y o n y que devuelva un nulo (NaN) si no se encuentra el valor.

fatal = df_archivo['fatal_(y/n)']
fatal.unique()
```

```python
def limpieza(column):
    if column == 'Y' or column == 'y':
        return 'Y'
    elif column == 'N' or column == ' N' or column == 'N ':
        return 'N'
    else:
        return np.NaN
```

```python
df_archivo['fatal_clean'] = df_archivo['fatal_(y/n)'].apply(limpieza)
df_archivo.sample()
```

```python
# Seguid la misma lógica que en el ejercicio anterior pero con la columna sex. 
# Pero en este caso vamos a incluir un retito más: En este caso querremos tener dos valores únicos F y M. 
# De nuevo en la columna hay otros carácteres raros como 'M ', 'lli', 'N', '.'. 
# En esos casos, queremos que se reemplacen por M o F de forma aleatoria (tendréis que usar la librería random). 
# En caso de que sea nulo, que la función devuelva nulo (np.nan).

sexo = df_archivo['sex_']
sexo.unique()
```

```python
def limpieza_2(column):
    lista = ['F', 'M']
    if column == 'F':
        return 'F'
    elif column == 'M' or column == 'M ' or column == 'N':
        return 'M'
    elif column == 'lli' or column == '.':
        return random.choice(lista)
    else:
        return np.NaN
```

```python
df_archivo['sexo'] = df_archivo['sex_'].apply(limpieza_2)
df_archivo['sexo'].unique()
```
