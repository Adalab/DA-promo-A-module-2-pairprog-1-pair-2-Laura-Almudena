```python
# porcentaje de valores nulos

casos_nulos = archivo_df.isnull().sum()
print(casos_nulos)

casos_totales = archivo_df.shape[0]         


resultado = (casos_nulos * 100) / casos_totales
print(resultado)


# % -->
#  casos total ----- 100 %
#  casos que tengo --- x % --->  X = (casos nulos * 100%) / casos totales
```

```python
# Sacad los valores únicos y su frecuencia de las variables numéricas. 
# Almacenad los resultados en un diccionario. Uno para cada variable.                  

df_numericas = archivo_df[["Year", "original order", "Unnamed: 0"]]
dic = {}

for i in df_numericas:
    df_numericas[i].value_counts()
    dic[i] = df_numericas[i].value_counts()
```
