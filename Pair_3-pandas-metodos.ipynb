```python
######################################################### FALTA sacar las variables por separado
# separa variables categóricas y numéricas
# archivo_df.info()

# object es categórico y el resto: float64 e int64 numérico
tipos_arc = archivo_df.dtypes
# print(tipos_arc)

numericas = []
categoricas = []

for x in range(archivo_df.dtypes.size):

    if tipos_arc[:1]== "int64":
        numericas.append("int64")
    elif tipos_arc[:1] == "float64":
        numericas.append("float64")
    else: 
        categoricas.append("object")

print(numericas)
print(categoricas)


# corte_df = archivo_df.cut(archivo_df.dtypes, 2)
```

```python
# porcentaje de valores nulos

casos_nulos = archivo_df.isnull().sum()
print(casos_nulos)

casos_totales = archivo_df.shape[0]         


resultado = (casos_nulos * 100) / casos_totales
print(resultado)


# % -->
#  casos total ----- 100 %
#  casos que tengo --- x % --->  X = (casos que tengo * 100%) / casos totales
```
